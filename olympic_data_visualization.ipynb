### Missing Values

```python
n_records = len(athletes)
def missing_values_df(df):
    for column in df:
        print("{} | {} | {}".format(
            column, len(df[df[column].isnull()]) / (1.0*n_records), df[column].dtype
        ))

missing_values_df(athletes)
```

### Variable Grouping

```python
height_types = ['very short', "short",  "medium", "tall", "very tall"]

male_athletes['height_type'] = pd.qcut(male_athletes['Height'],
                                  5, height_types)
```

```python
male_athletes.head()
```

```python
height_types = ['very short', "short",  "medium", "tall", "very tall"]

female_athletes['height_type'] = pd.qcut(female_athletes['Height'],
                                  5, height_types)
```

```python
weight_types = ['very light', "light",  "medium", "heavy", "very heavy"]

male_athletes['weight_types'] = pd.qcut(male_athletes['Weight'],
                                  5, weight_types)
```

```python
weight_types = ['very light', "light",  "medium", "heavy", "very heavy"]

female_athletes['weight_types'] = pd.qcut(female_athletes['Weight'],
                                  5, weight_types)
```

# Data Visualization

### Note: height is in cm and weight is in kg

```python
athletes = pd.read_csv("data/athletes_clean.csv")
male_athletes = pd.read_csv("data/male_athletes_clean.csv")
female_athletes = pd.read_csv("data/female_athletes_clean.csv")
```

## Changes Over Time

```python
male_athletes.groupby('Year').Height.mean().plot.line();
```

^ year vs height for males

```python
female_athletes.groupby('Year').Height.mean().plot.line();
```

^ year vs height for females

```python
male_athletes.groupby('Year').Weight.mean().plot.line();
```

^ year vs weight for males

```python
female_athletes.groupby('Year').Weight.mean().plot.line();
```

^ year vs weight for females

### Athletes have gotten taller and heavier over time!

### Now lets look at Summer vs Winter Olympics

```python
male_summer = male_athletes[male_athletes.Season == 'Summer']
male_winter = male_athletes[male_athletes.Season == 'Winter']
female_summer = female_athletes[female_athletes.Season == 'Summer']
female_winter = female_athletes[female_athletes.Season == 'Winter']
```

```python
ax = male_summer.groupby('Year').Height.mean().plot.line();
male_winter.groupby('Year').Height.mean().plot.line(ax = ax)
```

^ year vs height for males, where blue is Summer and orange is Winter 

```python
ax = female_summer.groupby('Year').Height.mean().plot.line();
female_winter.groupby('Year').Height.mean().plot.line(ax = ax)
```

^ year vs height for females, where blue is Summer and orange is Winter 

```python
ax = male_summer.groupby('Year').Weight.mean().plot.line();
male_winter.groupby('Year').Weight.mean().plot.line(ax = ax)
```

^ year vs weight for males, where blue is Summer and orange is Winter 

```python
ax = female_summer.groupby('Year').Weight.mean().plot.line();
female_winter.groupby('Year').Weight.mean().plot.line(ax = ax)
```

^ year vs weight for females, where blue is Summer and orange is Winter 

### So athletes are generally taller in the Summer Olympics. Athletes have typically weighed about the same in the Summer and Winter Olympics, until very recently where athletes have weighed more in the Summer. Interesting!

```python
female_percentage = female_athletes.groupby('Year').count()['ID'] / athletes.groupby('Year').count()['ID']
female_percentage.plot()
```

^ year vs percentage of woman athletes

### The number of women athletes has gone up but we're still not quite at 50 percent
